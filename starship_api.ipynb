```python
import requests
import json
from pprint import pp
```

```python
url = 'https://swapi.dev/api/starships?page=1'
results_data = []
while url:
    page_content_dict = requests.get(url).json()
    results_data.extend(page_content_dict["results"])
    url = page_content_dict["next"]
    print(url)
pp(results_data)
```

```python
#Write a function that returns all of the models of starships
def get_models():
    model_list=[item['model'] for item in results_data]
    return model_list
   
pp(get_models())
```

```python
#Write a function that finds the ship that can carry the most cargo
def get_most_cargo():
    cargo_dict={}
    #get ship name and cargo_capacity in a dictionary
    cargo_dict={item['name']:int(item['cargo_capacity']) for item in results_data if (item['cargo_capacity'].isnumeric())}
    # get max cargo capacity
    max_cargo = max(cargo_dict.values())
    # cargo_dict=dict(sorted(cargo_dict.items(), key=lambda kv:(kv[1], kv[0]),reverse=True)) 
    #get ship name for the max cargo
    ship_name=[key for key in cargo_dict if cargo_dict[key]==max_cargo]
    return ship_name
    
print("Ship that can carry most cargo '{0}'".format(get_most_cargo()[0]))
```

```python
#Write a function that fings of crew and passengers on a given starship
def find_crew_passenger(starship):

    ship={item['crew']:item['passengers'] for item in  results_data if item['name']==starship}
    return list(map(list,ship.items()))

ship_name=input("Enter the ship name: ")
result=find_crew_passenger(ship_name)
print(f"There are '{result[0][0]}' crew members and '{result[0][1]}' passengers in '{ship_name}'")
```

```python
#Write a function to find the most expensive starship
def expensive_starship():
    costly_ship={}
     #get ship name and cost in a dictionary key value pair
    costly_ship={item['name']:int(item['cost_in_credits']) for item in results_data if (item['cost_in_credits'].isnumeric())}
    max_exp = max(costly_ship.values())
    #costly_ship=dict(sorted(costly_ship.items(), key=lambda kv:(kv[1], kv[0]),reverse=True)) 
    ship_name=[key for key in costly_ship if costly_ship[key]==max_exp]
    return ship_name


expensive_starship()       
print("Ship that can carry most cargo '{0}'".format(expensive_starship()[0]))
```

```python
#Write a function to find all of the starships that are less than a given price
def starship_lessthan_price(price):
    ship_name=[item['name'] for item in results_data if item['cost_in_credits'].isnumeric() and (int(item['cost_in_credits']) < price)]
    return ship_name

ship_price=int(input("Enter ship price:"))
print(f"Starship that are less than {ship_price} are",starship_lessthan_price(ship_price))
```

```python
#Write a function to find the starships that have appeared in a given number of films
def starship_films(num):
    ship_name=[item['name'] for item in results_data if len(item['films']) == num]
    return ship_name

film_num=int(input("Enter number of films:"))
print(f"starships that have appeared in {film_num} number of films are",starship_films(3))
```
